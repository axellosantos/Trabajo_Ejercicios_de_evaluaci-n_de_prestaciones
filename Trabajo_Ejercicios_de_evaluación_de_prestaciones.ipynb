
EJERCICIO-1
=======
Se pretende ejecutar una aplicación en un computador cuyo procesador emplea únicamente dos tipos de instrucción:

	 	Instrucción 1 con CPI = 3
	 
	 	Instrucción 2 con CPI = 2

Se sabe además que el procesador opera a 200 MHz.

Para generar el ejecutable del programa se pueden emplear dos compiladores diferentes, A y B, que ofrecen dos interpretaciones distintas del código, como indica la tabla adjunta.

 | Compilador | Instrucción 1| Instrucción 2 |
 | ---------- | --------     | ------------------ 
 | A          | 6            | 2   
 | B          | 4            | 6      


  ##	1.1 Obtener los MIPS necesarios para ejecutar ambas opciones en la máquina indicada.


Se sabe que I1 con CPI 3 tiene 6 instrucciones, por lo tanto 6 x 3 = **18 ciclos**  
Se sabe que I2 con CPI 2 tiene 2 instrucciones, por lo tanto 2 x 2 = **4 ciclos**  
Y que la frecuencia son **200 Mhz** = 100 *106 Hz.  

Sabiendo que:



$$
CPI = \frac{\sum Ni \times CPI}{Ni}
$$

Podemos hallar los ciclos por instrucción.





$$
\frac{18+4}{6+2} = \frac{22}{8} \textrm{ciclos de instruccion}
$$


Para calcular los MIPS usaremos la siguiente expresión matemática. 


$$
MIPS = \frac{F}{CPI * 10^{6}}
$$















Sustituyendo nos queda:

$$
MIPS = \frac{200 \times 10^{6}}{\frac{22}{8} \times 10^6} = 72,\widehat{73} \textrm{ millones de intrucciones por segundo}
$$






Repetimo el mismo prodeciiento para B
Se sabe que I1 con CPI 3 tiene 4 instrucciones, por lo tanto 3 x 4 = 12 ciclos
Se sabe que I2 con CPI 2 tiene 6 instrucciones, por lo tanto 6 x 2 = 12 ciclos


Podemos hallar los ciclos por instrucción :



$$
\frac{12 + 12}{6 + 4} = \frac{24}{10} \textrm{ciclos de instruccion}
$$

Usando la expresión de arriba, obtenemos los MIPS en el caso B





$$
MIPS = \frac{200 \times 10^{6}}{\frac{24}{10} \times 10^6} = 83,\widehat{3} \textrm{ millones de intrucciones por segundo}
$$



  
  
  
  

## 1.2 Teniendo en cuenta el tiempo de CPU, compara y comenta los resultados obtenidos en el apartado anterior.

Primero hallamos los tiempos de CPU de ambas. Utilizaremos la siguiente expresión matemática:
$$
 TcpuA = \frac{Ni \times CPI}{F} => \frac{18 \times \frac{22}{8} }{200 \times 10^{6}} = 2.47 \times 10^{-7} s
$$


$$
 TcpuB = \frac{Ni \times CPI}{F} => \frac{4 \times \frac{24}{10} }{200 \times 10^{6}} = 4.8 \times 10^{-8} s 
$$





























EJERCICIO-2
=======

Durante el desarrollo de una CPU se evalúa su rendimiento mediante un **benchmark** y, entre otros resultados, se obtiene la siguiente distribución del tiempo de ejecución:  

 | Instrucción 		| %         | 
 | ---------- 		| --------   | 
 | Multiplicación   | 20          | 
 | Acceso a memoria | 50          |      
 | Resto          	| 30          |   
 


En función de estos resultados se estudia introducir dos opciones de mejora en la configuración: 
   
	Opción 1: Acelerar en 4 veces la ejecución de multiplicaciones.
	Opción 2: Mejorar el acceso a operandos en memoria, reduciendo a la mitad el tiempo necesario para ello.


Se pide:  
## 2.1 	¿Cuál es la aceleración si solo se aplica la opción 1?


Utilizando la Ley de Amdahn, que dice:

$$
A = \frac{ \textrm{ Tiempo sin mejora}}{\textrm{ Tiempo con mejora}} = \frac{1}{(1 - F_{m}) + \frac{F_{m}}{A_{m}}}
$$

Hallamos la aceleración para Opcion 1 sabiendo que 
$$
F_{m} = 20\% 
$$

$$
Am = 4
$$

$$
A = \frac{1}{ (1 - 20\%) + \frac{20\%}{4} } = 1.17
$$



## 2.2 ¿Y si solo se aplica la opción 2?  
$$
F_{m} = 50\% 
$$

$$
Am = 2
$$

$$
A = \frac{1}{ (1 - 50\%) + \frac{50\%}{2} } = 1.\widehat{3}
$$

## 2.3 Evalúa de nuevo la aceleración si se aplican las dos opciones planteadas.

Mediante la misma ley , aplicando las opciones planteadas, el resultado seria:


$$
A = \frac{1}{ (1 - 70\%) + \frac{50\%}{2}  + \frac{20\%}{4} } = 1.73
$$
